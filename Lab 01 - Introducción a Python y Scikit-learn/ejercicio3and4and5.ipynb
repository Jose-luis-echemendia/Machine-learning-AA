```python
import pandas as pd

# 3. Crear el diccionario con los datos
datos = {
    'ciudad': ['La Habana', 'Madrid', 'Ciudad de México', 'Santiago de Cuba', 'Barcelona'],
    'población': [2100000, 3200000, 9200000, 430000, 1600000],
    'país': ['Cuba', 'España', 'México', 'Cuba', 'España']
}

# Convertir el diccionario a un DataFrame de pandas
df = pd.DataFrame(datos)

print("--- DataFrame Original ---")
print(df)
print("\n" + "="*30 + "\n")

# 3.a) Filtrar para mostrar solo las ciudades de un país (ej. 'Cuba')
print("--- a) Ciudades pertenecientes a 'Cuba' ---")
ciudades_cuba = df[df['país'] == 'Cuba']
print(ciudades_cuba)
print("\n" + "="*30 + "\n")


# 3.b) Contar cuántas ciudades tienen más de 500,000 habitantes
# Primero, creamos una condición booleana
condicion_poblacion = df['población'] > 500000

# Luego, sumamos los resultados (True=1, False=0) para obtener el conteo
cantidad_ciudades_grandes = condicion_poblacion.sum()

print("--- b) Conteo de ciudades con más de 500,000 habitantes ---")
print(f"Número de ciudades con más de 500,000 habitantes: {cantidad_ciudades_grandes}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reutilizamos los datos y el DataFrame del ejercicio anterior
datos = {
    'ciudad': ['La Habana', 'Madrid', 'Ciudad de México', 'Santiago de Cuba', 'Barcelona'],
    'población': [2100000, 3200000, 9200000, 430000, 1600000],
    'país': ['Cuba', 'España', 'México', 'Cuba', 'España']
}
df = pd.DataFrame(datos)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6)) # Ajusta el tamaño para mejor visualización
plt.bar(df['ciudad'], df['población'], color=['blue', 'red', 'green', 'blue', 'red'])

# Añadir títulos y etiquetas para mayor claridad
plt.title('Población por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Población (en millones)')
plt.xticks(rotation=45, ha="right") # Rotar las etiquetas del eje x para que no se solapen
plt.tight_layout() # Ajusta el gráfico para que todo encaje bien

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un rango de valores para el eje X (por ejemplo, de -2*pi a 2*pi)
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calcular el valor de la tangente para cada punto de x
y = np.tan(x)

# Graficar la función
plt.figure(figsize=(10, 6))
plt.plot(x, y)

# Añadir títulos y etiquetas
plt.title("Gráfico de la Función Tangente (y = tan(x))")
plt.xlabel("x (radianes)")
plt.ylabel("tan(x)")

# Limitar el eje Y para evitar las asíntotas infinitas y hacer el gráfico legible
plt.ylim(-5, 5)

# Añadir una cuadrícula para facilitar la lectura
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Mostrar el gráfico
plt.show()
```
